```python
# high / low game
import random

levels = [
    {'min': 1, 'max': 10, 'tries': 3},
    {'min': 1, 'max': 100, 'tries': 5},
    {'min': 1, 'max': 1000, 'tries': 10}
]

def generate_answer(level):
    # generate a random number between min - max
    answer = random.randint(level['min'], level['max'])
    print(answer)
    return answer

def take_a_guess():
    # ask user to make a guess
    guess = int(input("Guess a number: "))
    return guess

def check_answer(answer, guess):
    if answer < guess:
        print("Too High")
        return False
    elif answer > guess:
        print("Too Low")
        return False
    else:
        return True
    
def check_for_winner(win):
    if win:
        print("You WIN!")
    else:
        print("You LOSE!")

def game_loop(answer, allowed_guesses):
    num_guesses = 0
    while num_guesses < allowed_guesses:
        guess = take_a_guess()
        win = check_answer(answer, guess)
        if win:
            break
        num_guesses += 1
    return [guess, win]

def high_low_game():
    allowed_guesses = 3
    current_level = 1
    choice = 'Y'
    while current_level <= 3 and choice == 'Y':
        print(f"Level: {current_level}")
        print(f"Range: {levels[current_level - 1]['min']} to {levels[current_level - 1]['max']}")
        answer = generate_answer(levels[current_level - 1])
        guess, win = game_loop(answer, allowed_guesses)
        check_for_winner(win)
        # check to see if user wants next level on win
        choice = 'N'
        if win:
            if current_level < 3:
                choice = input("Do you want to continue? (Y/N) ")
                if choice == 'Y':
                    current_level += 1
            else:
                print("Achievement Unlocked! MASTER LEVEL!")

# start game
high_low_game()
```

```python
# pass by reference
def pass_by_reference(my_list, my_dict):
    my_list[0] = 234
    my_list.append(999)
    my_dict['orange'] = 'orange'
    my_list = [6, 7, 8, 9, 10]
    print(f"inside: l = {my_list}, d = {my_dict}")

l = [1, 2, 3, 4, 5]
d = {'apple': 'red', 'banana': 'yellow'}
print(f"before: l = {l}, d = {d}")
pass_by_reference(l, d)
print(f"after: l = {l}, d = {d}")
```
