```python
import ipytest
ipytest.autoconfig()
```

```python
# a program that adds two numbers

def add(a, b):
    return a + b


# unit test to determine if program works
assert add(2, 3) == 5
```

```python
# create a program that asks for the user's age
# the program will return true, if the user is an adult

def get_user_age():
    return int(input("Enter age: ")) >= 18

assert get_user_age() == True
```

```python
# pytest basics
%%ipytest


# functions start with "test_"
# contains asserts that are testing a single aspect of the program
def test_number_assumptions():
    age = 15

    # check to see if the age is a particular value
    assert age == 25

    # check to see if the age is over 18
    assert age > 18

    # check to see if the age is under 100
    assert age < 100
```
